```python
# most pitchers to allow multiple runs in an inning on one team
```

```python
import pandas as pd
import boxball_loader as bbl
```

```python
id_cols = ['game_id', 'inn_ct', 'bat_home_id', 'event_id']
cols = id_cols + ['bat_id', 'resp_pit_id', 'bat_fate_id']
evs = bbl.load_event_data(bbl.Eras.ThirtyTeams, cols)[cols]
evs.shape
```

```python

evs
```

```python
runs_allowed = evs[evs['bat_fate_id'] > 3]
```

```python

by_inning = runs_allowed.groupby(['game_id', 'inn_ct', 'bat_home_id', 'resp_pit_id']).count().rename(columns={'event_id': 'runs_ct'}).reset_index()
by_inning
```

```python
by_inning[by_inning['runs_ct']>1]
```

```python
# most pitchers giving up at least one run in a half-inning
by_inning.groupby(['game_id', 'inn_ct', 'bat_home_id'])['resp_pit_id'].nunique().sort_values()
```

```python
# most pitchers giving up multiple runs in a half-inning
by_inning[by_inning['runs_ct']>1].groupby(['game_id', 'inn_ct', 'bat_home_id'])['resp_pit_id'].nunique().sort_values()
```
