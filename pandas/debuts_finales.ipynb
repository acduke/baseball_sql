```python
# MLB debuts seen

import boxball_loader as bbl
import baseball_stats_utils as bsu
```

```python
dailies = bbl.load_dailies(game_types=bbl.GameType.ALL).sort_values(by=['game_dt', 'game_ct'])
```

```python
cols = ['player_id', 'team_id', 'game_dt', 'game_ct', 'game_id']
debuts = dailies.groupby('player_id').head(1)[cols].set_index('player_id')
debuts['name'] = bsu.get_player_names_df(debuts, 'retro_id')
finales = dailies.groupby('player_id').tail(1)[cols].set_index('player_id')
finales['name'] = bsu.get_player_names_df(finales, 'retro_id')
```

```python
debuts.to_csv('debuts.csv')
finales.to_csv('finales.csv')
```

```python
with open('known_games.txt') as f:
    known_games = [gm.rstrip() for gm in f.readlines()]

known_games
```
