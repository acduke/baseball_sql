```python
import pandas as pd
import pyreadr
import plotly.express as px
```

```python
result = pyreadr.read_r('~/Downloads/MLBStadiaPathData.rda')
df = result['MLBStadiumsPathData']
df
```

```python
angels = df[df['team']=='angels']
angels
```

```python
fig = px.scatter(angels, x='x', y='y', color='segment')
fig.update_layout(
    xaxis_range=[0, 250], yaxis_range=[0, 250],
    autosize=False,
    width=600,
    height=500)
fig.update_traces(marker={'size': 3})
fig.show()
```
