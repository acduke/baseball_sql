### MLB Stadium Paths

There is an R project at https://github.com/bdilday/GeomMLBStadiums that includes stadium layout data, for visualization such as spray charts.  This reads the data in python.

```python
import pandas as pd
import pyreadr
import plotly.express as px
```

```python
# From https://github.com/bdilday/GeomMLBStadiums
result = pyreadr.read_r('../data/GeomMLBStadiums/MLBStadiaPathData.rda')
df = result['MLBStadiumsPathData']
df
```

```python
angels = df[df['team']=='angels']
angels
```

```python
fig = px.line(angels, x='x', y='y', color='segment')
fig.update_layout(
    xaxis_range=[0, 250], yaxis_range=[0, 250],
    autosize=False,
    width=600,
    height=500)

fig.show()
```

```python
def draw_park(name):
    park = df.query('team==@name')
    fig = px.line(park, x='x', y='y', color='segment')
    fig.update_layout(
        xaxis_range=[0, 250], yaxis_range=[0, 250],
        autosize=False,
        width=600,
        height=500)

    return fig

draw_park('padres').show()
```

```python
for park in df['team'].drop_duplicates():
    print(park)
    draw_park(park).show()
```
