```python
import boxball_loader as bbl
import pandas as pd
```

```python
hof = pd.read_csv('../data/bbref/hof.csv')
hof
```

```python
hof['bbref_id'] = hof['Name'].apply(lambda x: x.split("\\")[1])
hof
```

```python
ppl = bbl.load_people()[['bbref_id', 'retro_id']]
hofers = pd.merge(left=hof, right=ppl, on='bbref_id')['retro_id'].dropna().values
hofers
```

```python
glt = bbl.load_gamelog_teams(game_types=(bbl.GameType.RS|bbl.GameType.PS), years=range(1800, 3000))
glt.columns
```

```python
hof_starts = glt.query('starting_pitcher_id in @hofers')[['game_id', 'team', 'starting_pitcher_id']]
hof_starts
```

```python
catchers = bbl.load_gamelog_starters(game_types=bbl.GameType.ALL, years=range(1800, 3000)).query("pos==2")
catchers.sample(10)
```

```python
starts = pd.merge(left=hof_starts, right=catchers, on=['game_id', 'team']).rename(columns={'player_id': 'catcher_id'})
starts
```

```python
starts[['starting_pitcher_id', 'catcher_id']].drop_duplicates()['catcher_id'].value_counts().head(10)
```

```python
starts[['starting_pitcher_id', 'catcher_id']].query('catcher_id=="davij001"').value_counts()
```

```python
ev = bbl.load_event_data(1800, 3000, ['pit_id', 'pos2_fld_id'], game_types=(bbl.GameType.ALL))
ev.sample(10)
```
