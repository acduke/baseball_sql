# Why did the White Sox have such a great winning percentage when Mike Squires played?

Effectively Wild looked at teams' W-L record with and without individual players, and found that Mike Squires consistently had a big W-L WOWY split.  Why is that?

https://blogs.fangraphs.com/effectively-wild-episode-1842-do-not-pass-first-do-not-collect-a-run/

```python
import pandas as pd
import boxball_loader as bbl
```

```python
df = bbl.load_dailies(bbl.GameType.RS)
df.shape
```

```python
df.columns
```

```python
squires = df.query('1979 <= yr <= 1984').query('player_id=="squim001"')
squires
```

```python
squires.columns.values
```

```python
squires['seq_ct'].value_counts()
```

```python
glt = bbl.load_gamelog_teams(game_types=bbl.GameType.RS, years=range(1982, 1985))
```

```python
glt.query('team=="CHA"')
```

```python
squires_starts = squires.query('seq_ct==1')['game_id']
squires_later = squires.query('seq_ct>1')['game_id']
```

```python
glt.query('team=="CHA"').query('game_id in @squires_starts')[['W', 'L']].sum()
```

```python
glt.query('team=="CHA"').query('game_id in @squires_later')[['W', 'L']].sum()
```

```python
glt.query('team=="CHA"').query('game_id not in @squires_starts')[['W', 'L']].sum()
```
