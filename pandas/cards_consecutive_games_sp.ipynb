```python
import pandas as pd

df = pd.read_parquet('../data/mine/gl_teams.parquet')
```

```python
# Get all regular-season Cardinals games
cards = df[(df['team']=='SLN')&(df['game_type']=='RS')]
sps = cards[['yr', 'date', 'team_game_number', 'starting_pitcher_id']].sort_values(by=['yr', 'team_game_number'])

# Find the streaks
sps['streak_id'] = sps['team_game_number'] - sps.groupby('starting_pitcher_id').cumcount()
streaks = sps.groupby(['starting_pitcher_id', 'yr', 'streak_id']).filter(lambda group: len(group) > 1)

streaks.tail(50)
```
