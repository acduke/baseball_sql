```python
import pandas as pd
import numpy as np 
import boxball_loader as bbl
```

```python
# select the rows and columns we're interested in
dailies = bbl.load_dailies(game_types=(bbl.GameType.RS|bbl.GameType.PS)).query('yr==2019')[['game_id', 'player_id', 'b_g_dh']]
dailies
```

```python
# for each game, count the number of players used and whether anybody was a DH
num_players = dailies.groupby('game_id').agg(num_players=('player_id', np.size), dh_fl=('b_g_dh', max))
num_players
```
