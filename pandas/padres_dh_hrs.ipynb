### First Padre to homer as a DH

On the Padres' broadcast on August 16, 2020, the trivia question asked who was the first Padre to HR as a DH.

They implicitly limited it to the regular season, as Kurt Bevacqua homered as a DH in the '84 WS, but wasn't one of the choices (the choices were Rickey Henderson, Tony Gwynn, Greg Vaughn and Mark Sweeney).

```python
import pandas as pd
import boxball_loader as bbl
```

```python
df = bbl.load_event_data(1995, 3000, ['bat_fld_cd', 'away_team_id', 'bat_home_id', 'bat_id', 'inn_ct'])
pads_dh_hr = df.query('bat_fld_cd==10 and away_team_id=="SDN" and event_cd==23 and not bat_home_id')[['date', 'game_id', 'bat_id', 'inn_ct']]
pads_dh_hr
```
