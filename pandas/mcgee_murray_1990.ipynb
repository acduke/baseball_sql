# McGee vs Murray, 1990

What did the NL BA leaderboard look like when McGee was traded to the A's on August 30, 1990?

```python
import boxball_loader as bbl
import pandas as pd
```

```python
df = bbl.load_dailies_bat(game_types=bbl.GameType.RS).query('yr==1990')
```

```python
aug = df[(df['game_dt']< '1990-08-31')].groupby('player_id')[['b_pa', 'b_ab', 'b_h']].sum()
aug
```

```python
aug['ba'] = aug['b_h']/aug['b_ab']
```

```python
aug[aug['b_pa']>= 120*3.1].sort_values('ba', ascending=False).head(25)
```
