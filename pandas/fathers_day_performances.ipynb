```python
import pandas as pd
import datetime

# Father's Day is the third Sunday in June 
def get_fathers_day_yr(yr):
    june_1st_day = datetime.date(yr,6,1).weekday()
    return datetime.date(yr, 6, 21-june_1st_day)

(get_fathers_day_yr(2019), get_fathers_day_yr(2020), get_fathers_day_yr(2036))
```

```python
fathers_days = [get_fathers_day_yr(yr) for yr in range(1910,2020)]
fathers_days
```

```python
df = pd.read_parquet("file:daily.parquet").drop_duplicates(['game_id', 'player_id'])

fdays = df[df['game_dt'].isin(fathers_days)]
fdays
```

```python
pitching_columns=['player_id', 'p_g', 'p_gs', 'p_out', 'p_er', 'p_w', 'p_l', 'p_so']
fdays_pitching = fdays[fdays['p_g']==1][pitching_columns].groupby('player_id').sum().sort_values(by='p_gs', ascending=False).head(25)
fdays_pitching['era'] = fdays_pitching['p_er']/fdays_pitching['p_out']*27
fdays_pitching
```
