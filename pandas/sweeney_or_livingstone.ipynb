```python
import pandas as pd
import boxball_loader as bbl
```

```python
dailies = bbl.load_dailies(game_types=bbl.GameType.ALL)
dailies
```

```python
def get_appearances(player_id):
    return dailies.query('player_id==@player_id & team_id=="SDN" & opponent_id=="LAN" & not home_fl')[['game_id', 'game_dt', 'slot_ct', 'seq_ct']] 

appearances = {player_id: get_appearances(player_id) for player_id in ['sweem001', 'gwynt001', 'livis001']}

appearances['livis001']
```

```python
pd.merge(appearances['sweem001'], appearances['gwynt001'], on ='game_id')
```

```python
pd.merge(appearances['livis001'], appearances['gwynt001'], on ='game_id')
```
