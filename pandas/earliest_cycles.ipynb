```python
import boxball_loader as bbl
import baseball_stats_utils as bsu
```

```python
df = bbl.load_dailies_bat(bbl.GameType.RS)
df.sample(10)
```

```python
df.columns
```

```python
df['b_1b'] = df['b_h'] - df['b_2b'] - df['b_3b'] - df['b_hr']
```

```python
cycles = df.query('b_1b>0 and b_2b>0 and b_3b>0 and b_hr>0')
cycles
```

```python
df['inc_h'] = df.sort_values(['player_id', 'yr', 'game_dt']).groupby(['player_id', 'yr'])['b_h'].cumsum()-df['b_h']
```
