```python
import pandas as pd
import numpy as np
import boxball_loader as bbl
```

```python
gl = bbl.load_gamelogs(bbl.GameType.RS, range(2012, 2021))
gl
```

```python
weekday_types = pd.Series({'Sat': 'we', 'Sun': 'we', 'Fri': 'n', 'Wed': 'wd', 'Tue': 'wd', 'Mon': 'wd', 'Thu': 'wd'}).rename('weekday_type')
weekday_types.index.name='day_of_week'
weekday_types
```

```python
glw = pd.merge(left=gl, right=weekday_types, on='day_of_week')
glw
```

```python
att = glw.groupby(['home_team', 'weekday_type']).agg({'attendance': [np.mean, min, max, len]})
att
```

```python
att_means = att['attendance']['mean'].unstack()
att_means
```

```python
# Ratio of weekday to weekend
(att_means['wd']/att_means['we']).sort_values()
```
