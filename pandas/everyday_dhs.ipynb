```python
import pandas as pd

import boxball_loader as bbl
```

```python
apps = bbl.load_appearances()
apps.sample(10)
```

```python
totals = apps.groupby('player_id')[['g_all', 'g_batting', 'g_defense', 'g_dh']].sum()
totals
```

```python
totals['dh_pct'] = totals['g_dh'] / totals['g_batting']
totals.query('g_batting > 100').sort_values('dh_pct')
```
